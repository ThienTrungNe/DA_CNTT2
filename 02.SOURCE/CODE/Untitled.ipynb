```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://tgroup.vn/travel-guide/giai-dap-cac-thac-mac-thuong-gap-ve-du-lich-da-lat")
soup = BeautifulSoup(response.content, "html.parser")
print(soup)
```

```python
titles = soup.findAll(class_='moto-widget-accordion__header-item moto-widget-accordion__header-title')
print(titles)
```

```python
links = [link.text for link in titles]
print(links)
```
